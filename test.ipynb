```python
import json
import PyPDF2

pdf_file = "example_pdf.pdf"

pdf = PyPDF2.PdfReader(pdf_file)

data = {}

# Extract title
first_page = pdf.pages[0] 
title_text = first_page.extract_text()

for line in title_text.split("\n"):
    if line.istitle():
        data['title'] = line
        break

# Extract headings and text
headings = {}
content = ""

# Get unique font sizes
sizes = set() 
for page in pdf.pages:
    text = page.extract_text()
    chars = page.g() 
    for size in chars:
        if size > 0:
            sizes.add(size)
            
sizes = list(sizes)
sizes.sort(reverse=True)

for page in pdf.pages:
    text = page.extract_text ()
    chars = page._get_fonts
    
    current_size = None
    
    for i, char in enumerate(text):
        if chars[i] == 0:
            current_size = chars[i+1]
            
        if current_size == sizes[0]:
            # Heading level 1
            if previous_heading:
                headings[previous_heading].append(current_heading)
            current_heading = "" + char 
            previous_heading = current_heading
            
        elif current_size == sizes[1]:
            # Heading level 2
            if current_heading:
                current_heading += char
                
        else:
            # Regular text
            content += char
            
headings[previous_heading] = content            
data['headings'] = headings
data['content'] = content

# Save to JSON
with open('output.json', 'w') as f:
    json.dump(data, f, indent=4)
```

```python
import pdfplumber
import json

pdf = pdfplumber.open('Example_pdf.pdf')

font_sizes = {}

for page in pdf.pages:

  chars = page.chars
 
  for font_char in chars:
    font = font_char["fontname"]
    font_size = font_char["size"]
    font_size = round(font_size, 1)
    
    if font_size not in font_sizes:
      font_sizes[font_size] = 0
      
    font_sizes[font_size] += 1  

pdf.close()

font_sizes_l = list(font_sizes)
font_sizes_l.sort(reverse=True)

occurence_font_sorted = dict(sorted(font_sizes.items(), key=lambda x: x[1], reverse=True))

font_dict = {}
font_dict[(primary_font := next(iter(occurence_font_sorted.keys())))] = {"hierarchy name":'p', 'hierarchy place':-1}
font_dict[font_sizes_l[0]] = {"hierarchy name":'Title', 'hierarchy place':0} if font_sizes_l[0] != font_dict[primary_font] else 0

for i, font in enumerate([f for f in font_sizes_l if f not in list(font_dict)]):
    font_dict[font] = {"hierarchy name":f'heading {i+1}', 'hierarchy place':i}
```

```python
import json
import pdfplumber

pdf = pdfplumber.open("Example_pdf.pdf")

paragraphs = []
headings = {}
data = {}
previous_font_size = None

for page in pdf.pages:

  chars = page.chars
 
  for word in chars:

    current_heading = None
    

    current_font_size = round(word['size'],1)
    current_hierarchy_place = font_dict[current_font_size]['hierarchy place']
    if current_font_size == previous_font_size or previous_font_size == None:
      # Paragraph text
      paragraphs.append(word['text'])

    else:
      
      
      data[previous_font_size] = "".join(paragraphs)

      paragraphs = []
      paragraphs.append(word['text'])
    
    previous_font_size = current_font_size
    previous_hierarchy_place = current_hierarchy_place
data[previous_font_size] = "".join(paragraphs)
with open('output.json', 'w') as f:
  json.dump(data, f, indent=4)
```
